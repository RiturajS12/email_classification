```python
all_message = email_data['Messages']
all_message = [message.lower() for message in all_message]
```

```python
all_message = [re.sub(r"[^a-zA-Z ]", "", message) for message in all_message]
```

```python
email_data['Messages'] = all_message
```

```python
email_data
```

```python
email_data['Labels'].value_counts()
```

```python
ham = email_data[email_data['Labels']=='ham']
spam = email_data[email_data['Labels']=='spam']
```

```python
(ham.shape[0],spam.shape[0])
```

```python
ham = ham.sample(1000)
```

```python
ham.shape[0]
```

```python
#spam = spam.sample(1000,replace=True)
```

```python
Updated_df = pd.concat([ham,spam])
```

```python
Updated_df
```

```python
Updated_df.duplicated().sum()
```

```python
Updated_df['Labels'] = Updated_df['Labels'].map({'spam':1,'ham':0})
```

```python
x = Updated_df['Messages']
y = Updated_df['Labels']
```

```python
Updated_df
```

```python
from sklearn.feature_extraction.text import  CountVectorizer , TfidfVectorizer

bow_obj = CountVectorizer(max_features=1500,stop_words='english')

X = bow_obj.fit_transform(x)
X
```

```python
X.toarray()
```

```python
X_transform = X.toarray()
```

```python
X_transform
```

```python
y.shape
```

```python
X_transform.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_transform, y, test_size=0.15, random_state=42)
```

```python
#for text classification naive byes are best for the performance
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
#if y = classes 2 then bernoulli if more than two then multinomial and discrete gaussian

mnb = MultinomialNB()
bnb = BernoulliNB()

mnb.fit(x_train,y_train)
bnb.fit(x_train,y_train)
```

```python
print("MNB training:",mnb.score(x_train,y_train)*100,"testing:",mnb.score(x_test,y_test)*100)
print("BNB training:",bnb.score(x_train,y_train)*100,"testing:",bnb.score(x_test,y_test)*100)
```

```python
prediction = mnb.predict(x_test)
prediction
```

```python
evaluation_df = pd.DataFrame({"actuall_class":y_test,"Predicted_class":prediction})
evaluation_df
```

```python
evaluation_df['actuall_class'] = evaluation_df['actuall_class'].map({1:'spam',0:'ham'})
evaluation_df['Predicted_class'] = evaluation_df['Predicted_class'].map({1:'spam',0:'ham'})
```

```python
evaluation_df.head(50)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(evaluation_df['actuall_class'], evaluation_df['Predicted_class'])
```

```python
cm
```

```python
sns.heatmap(cm,annot = True,fmt ='d')
plt.axis('off')
plt.show()
```

```python
from sklearn.metrics import classification_report


print(classification_report(evaluation_df['actuall_class'], evaluation_df['Predicted_class']))
```
